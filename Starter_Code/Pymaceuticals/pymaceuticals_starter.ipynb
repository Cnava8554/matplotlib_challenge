```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
mice_count = combined_df["Mouse ID"].nunique()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df.loc[combined_df.duplicated(subset = ['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice = combined_df.loc[combined_df['Mouse ID'] == 'g989', :]
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
remove_dup_df = combined_df[combined_df["Mouse ID"].isin(duplicate_mice)==False]
remove_dup_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count_2 = remove_dup_df["Mouse ID"].nunique()
mice_count_2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean = remove_dup_df["Tumor Volume (mm3)"].groupby(remove_dup_df["Drug Regimen"]).mean()
median = remove_dup_df["Tumor Volume (mm3)"].groupby(remove_dup_df["Drug Regimen"]).median()
var = remove_dup_df["Tumor Volume (mm3)"].groupby(remove_dup_df["Drug Regimen"]).var()
std = remove_dup_df["Tumor Volume (mm3)"].groupby(remove_dup_df["Drug Regimen"]).std()
sem = remove_dup_df["Tumor Volume (mm3)"].groupby(remove_dup_df["Drug Regimen"]).sem()

Summary_df = pd.DataFrame({"Mean Tumor Volume":mean, "Median Tumor Volume":median, "Variance of Tumor Volume":var, "Standard Deviation of Tumor Volume":std, "Standard Error of Tumor Volume":sem})
Summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.
summary_stats = remove_dup_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"])
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
capomulin_df = remove_dup_df.loc[remove_dup_df["Drug Regimen"]=="Capomulin",:]
ramicane_df = remove_dup_df.loc[remove_dup_df["Drug Regimen"]=="Ramicane",:]
infubinol_df = remove_dup_df.loc[remove_dup_df["Drug Regimen"]=="Infubinol",:]
ceftamin_df = remove_dup_df.loc[remove_dup_df["Drug Regimen"]=="Ceftamin",:]
# Start by getting the last (greatest) timepoint for each mouse
capomulin_last = capomulin_df.groupby("Mouse ID").max()["Timepoint"]
capomulin_vol = pd.DataFrame(capomulin_last)
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
capomulin_merge = pd.merge(capomulin_vol, remove_dup_df, on=("Mouse ID", "Timepoint"), how="left")
capomulin_merge.head()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
mouse_selection_df = capomulin_df.loc[capomulin_df["Mouse ID"] == "s185",:]
x_axis = mouse_selection_df["Timepoint"]
y_axis = mouse_selection_df["Tumor Volume (mm3)"]
figure1.ax1 = plt.subplots()
plt.title('Capomulin Treatment: Mouse s185')
plt.plot(x_axis, y_axis, linewidth=1, markersize=10, marker="o", color="green", label="Fahreneit")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
figure1, ax1 = plt.subplots()
average_volume_capomulin =capomulin_df.groupby(['Mouse ID']).mean()
marker_size=10
plt.scatter(average_volume_capomulin['Weight (g)'],average_volume_capomulin['Tumor Volume (mm3)'], color="green")
plt.title('Average Tumor Volume vs Mouse Weight')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Averag Tumor Volume (mm3)')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation_co = round(st.pearsonr(average_volume_capomulin["Weight (g)"], average_volume_capomulin["Tumor Volume (mm3)"])[0], 2)
print(f"The correlation between mouse weight and the average tumor volume is {correlation_co}")
```
